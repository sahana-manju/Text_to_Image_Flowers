```python
import json
import random
import shutil
import os
import numpy as np
import torch
import PIL
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2
```

```python
# Set a random seed for reproducibility
RANDOM_SEED = 42
random.seed(RANDOM_SEED)

# Paths to COCO directories and annotation files
train_images_dir = 'ML_Dataset/train2017/'
val_images_dir = 'ML_Dataset/val2017/'

train_ann_captions_file = 'ML_Dataset/annotations-3/captions_train2017.json'
train_ann_instances_file = 'ML_Dataset/annotations-3/instances_train2017.json'
train_ann_keypoints_file = 'ML_Dataset/annotations-3/person_keypoints_train2017.json'

val_ann_captions_file = 'ML_Dataset/annotations-3/captions_val2017.json'
val_ann_instances_file = 'ML_Dataset/annotations-3/instances_val2017.json'
val_ann_keypoints_file = 'ML_Dataset/annotations-3/person_keypoints_val2017.json'

# Output directories for subset
output_train_images_dir = 'ML_Dataset/subset_train2017/'
output_val_images_dir = 'ML_Dataset/subset_val2017/'

output_train_ann_captions_file = 'ML_Dataset/subset_captions_train2017.json'
output_val_ann_captions_file = 'ML_Dataset/subset_captions_val2017.json'

output_train_ann_instances_file = 'ML_Dataset/subset_instances_train2017.json'
output_val_ann_instances_file = 'ML_Dataset/subset_instances_val2017.json'

output_train_ann_keypoints_file = 'ML_Dataset/subset_keypoints_train2017.json'
output_val_ann_keypoints_file = 'ML_Dataset/subset_keypoints_val2017.json'
```

```python
# Create output directories if they don't exist
os.makedirs(output_train_images_dir, exist_ok=True)
os.makedirs(output_val_images_dir, exist_ok=True)
```

```python
def create_subset(ann_caption_file, ann_instance_file, ann_keypoints_file, image_dir, output_image_dir, output_ann_caption_file, output_ann_instance_file, output_ann_keypoints_file, num_images=5000):
    # Load the original caption annotations
    with open(ann_caption_file, 'r') as f:
        captions_data = json.load(f)
    
    # Load the original instance annotations
    with open(ann_instance_file, 'r') as f:
        instances_data = json.load(f)
    
    # Load the original keypoints annotations
    with open(ann_keypoints_file, 'r') as f:
        keypoints_data = json.load(f)

    # Extract all image IDs from captions
    all_image_ids = [img['image_id'] for img in captions_data['annotations']]

    # Randomly select the required number of images
    subset_image_ids = random.sample(all_image_ids, num_images)

    # Filter the annotations for the selected images
    subset_captions = [ann for ann in captions_data['annotations'] if ann['image_id'] in subset_image_ids]
    subset_images = [img for img in captions_data['images'] if img['id'] in subset_image_ids]

    # Filter the instance annotations for the selected images
    subset_instances = [ann for ann in instances_data['annotations'] if ann['image_id'] in subset_image_ids]

    # Filter the keypoints annotations for the selected images
    subset_keypoints = [ann for ann in keypoints_data['annotations'] if ann['image_id'] in subset_image_ids]

    # Create new JSON structures for the subsets
    subset_instances_data = {
        'info': instances_data['info'],
        'licenses': instances_data['licenses'],
        'images': subset_images,
        'annotations': subset_instances,
        'categories': instances_data['categories']
    }
    
    subset_captions_data = {
        'info': captions_data['info'],
        'licenses': captions_data['licenses'],
        'images': subset_images,
        'annotations': subset_captions
    }

    subset_keypoints_data = {
        'info': keypoints_data['info'],
        'licenses': keypoints_data['licenses'],
        'images': subset_images,
        'annotations': subset_keypoints,
        'categories': keypoints_data['categories']
    }

    # Copy the selected images to the new directory
    for image in subset_images:
        img_filename = image['file_name']
        src_img_path = os.path.join(image_dir, img_filename)
        dest_img_path = os.path.join(output_image_dir, img_filename)
        shutil.copy(src_img_path, dest_img_path)

    # Save the subset of instance annotations
    with open(output_ann_instance_file, 'w') as f:
        json.dump(subset_instances_data, f)

    # Save the subset of caption annotations
    with open(output_ann_caption_file, 'w') as f:
        json.dump(subset_captions_data, f)

    # Save the subset of keypoints annotations
    with open(output_ann_keypoints_file, 'w') as f:
        json.dump(subset_keypoints_data, f)
```

```python
# Create a subset of 5000 training images with associated captions, instances, and keypoints
create_subset(train_ann_captions_file, train_ann_instances_file, train_ann_keypoints_file, train_images_dir, output_train_images_dir, output_train_ann_captions_file, output_train_ann_instances_file, output_train_ann_keypoints_file, num_images=5000)

# Create a subset of 500 validation images with associated captions, instances, and keypoints
create_subset(val_ann_captions_file, val_ann_instances_file, val_ann_keypoints_file, val_images_dir, output_val_images_dir, output_val_ann_captions_file, output_val_ann_instances_file, output_val_ann_keypoints_file, num_images=500)
```

```python
# Load the image
image_dir = 'ML_Dataset/subset_val2017/'
image_filename = '000000000139.jpg'
image_path = os.path.join(image_dir, image_filename)
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB

# Load annotations
with open('ML_Dataset/subset_instances_val2017.json', 'r') as f:
    instances_data = json.load(f)
with open('ML_Dataset/subset_captions_val2017.json', 'r') as f:
    captions_data = json.load(f)
with open('ML_Dataset/subset_keypoints_val2017.json', 'r') as f:
    keypoints_data = json.load(f)

# Image ID corresponding to '000000000139.jpg'
image_id = 139

# Get annotations for this image
instance_ann = [ann for ann in instances_data['annotations'] if ann['image_id'] == image_id]
caption_ann = [ann for ann in captions_data['annotations'] if ann['image_id'] == image_id]
keypoint_ann = [ann for ann in keypoints_data['annotations'] if ann['image_id'] == image_id]

# Plot the image and annotations
plt.figure(figsize=(10, 10))
plt.imshow(img_rgb)

# Draw bounding boxes
for ann in instance_ann:
    bbox = ann['bbox']
    x, y, w, h = bbox
    rect = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='r', facecolor='none')
    plt.gca().add_patch(rect)
    category_name = [cat['name'] for cat in instances_data['categories'] if cat['id'] == ann['category_id']][0]
    plt.text(x, y - 5, category_name, color='red', fontsize=12)

# Plot keypoints
for kp_ann in keypoint_ann:
    keypoints = kp_ann['keypoints']
    for i in range(0, len(keypoints), 3):
        x_kp, y_kp, visibility = keypoints[i:i+3]
        if visibility > 0:
            plt.scatter(x_kp, y_kp, color='blue', s=50)

# Show captions
for cap in caption_ann:
    plt.text(10, img.shape[0] - 30, cap['caption'], color='white', fontsize=14, backgroundcolor='black')

plt.axis('off')
plt.show()
```
